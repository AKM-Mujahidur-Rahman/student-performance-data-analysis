```python
print("\n=== OUTLIERS (All Numeric Columns) ===")
numeric_cols = df.select_dtypes(include=[np.number]).columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).sum()
print(outliers)
```

```python
np.random.seed(42)

n_samples = len(df)
indices = np.random.permutation(n_samples)
split_idx = int(0.75 * n_samples)

train_idx = indices[:split_idx]
test_idx = indices[split_idx:]

train_df_manual = df.iloc[train_idx]
test_df_manual = df.iloc[test_idx]

print(f"Training set size: {len(train_df_manual)} ({100*len(train_df_manual)/len(df):.0f}%)")
print(f"Test set size: {len(test_df_manual)} ({100*len(test_df_manual)/len(df):.0f}%)")
```

```python
# Remove outliers function
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

# FIXED: Apply to the correct dataframes
train = remove_outliers(train_df_manual, 'Hours_Studied')
train = remove_outliers(train, 'Exam_Score') 

test = remove_outliers(test_df_manual, 'Hours_Studied')
test = remove_outliers(test, 'Exam_Score')  

print('Train shape after cleaning:', train.shape)
print('Test shape after cleaning: ', test.shape)
```

```python
features = ['Hours_Studied']

X_train = train[features].values   # shape: (n_train, 1)
y_train = train['Exam_Score'].values     # shape: (n_train,)

X_test = test[features].values
y_test = test['Exam_Score'].values

print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
print('X_test shape:', X_test.shape)
print('y_test shape:', y_test.shape)
```

```python
# Add bias column (column of ones)
def add_bias(X):
    ones = np.ones((X.shape[0], 1))
    return np.hstack([ones, X])

A_train = add_bias(X_train)
A_test = add_bias(X_test)

print(f"A_train shape: {A_train.shape}")
print(f"First row of A_train: {A_train[0]}")
```

```python
AT = A_train.T                          # A transpose
ATA = AT @ A_train                       # A'A
ATA_inv = np.linalg.inv(ATA)             # (A'A)^-1
ATy = AT @ y_train                       # A'y

beta = ATA_inv @ ATy                     # final coefficients

print('Intercept (beta_0):              ', round(beta[0], 2))
print('Coefficient for Hours_Studied:   ', round(beta[1], 2))
print(f"\nEquation: Exam_Score = {beta[0]:.2f} + {beta[1]:.2f} × Hours_Studied")
```

```python
y_pred_train = A_train @ beta

# look at first 5 actual vs predicted prices
for i in range(5):
    print(f'Actual: {y_train[i]:.0f}   Predicted: {y_pred_train[i]:.0f}')
```

```python
y_pred_test = A_test @ beta

# MSE
errors      = y_test - y_pred_test
squared     = errors ** 2
mse         = squared.mean()
rmse        = mse ** 0.5  

print('Test MSE: ', round(mse, 2))
print('Test RMSE:', round(rmse, 2), '(marks)')
```

```python

plt.figure(figsize=(7, 5))
plt.scatter(y_test, y_pred_test, alpha=0.3, color='steelblue')


min_val = min(y_test.min(), y_pred_test.min())
max_val = max(y_test.max(), y_pred_test.max())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linewidth=1.5)

plt.xlabel('Actual Exam Score')
plt.ylabel('Predicted Exam Score')
plt.title('Actual vs Predicted Exam Score (Test Set)')
plt.tight_layout()
plt.show()
```

```python
# categorical columns we want to include
cat_cols = ['School_Type', 'Parental_Involvement', 'Access_to_Resources']

# one-hot encode
train_dummies = pd.get_dummies(train[cat_cols], drop_first=True)
test_dummies  = pd.get_dummies(test[cat_cols],  drop_first=True)


# combine numerical features with the new dummy columns
X_train_full = np.hstack([X_train, train_dummies.values.astype(float)])
X_test_full  = np.hstack([X_test,  test_dummies.values.astype(float)])

# add bias column
A_train_full = np.hstack([np.ones((X_train_full.shape[0], 1)), X_train_full])
A_test_full  = np.hstack([np.ones((X_test_full.shape[0],  1)), X_test_full])

# normal equation
beta_full = np.linalg.inv(A_train_full.T @ A_train_full) @ A_train_full.T @ y_train

# predict and evaluate
y_pred_full = A_test_full @ beta_full
mse_full    = ((y_test - y_pred_full) ** 2).mean()
rmse_full   = mse_full ** 0.5

print('With categorical features — Test RMSE:', round(rmse_full, 2), '(marks)')
print('Without categorical features — Test RMSE:', round(rmse, 2), '(marks)')
```
